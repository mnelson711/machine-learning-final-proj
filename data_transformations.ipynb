```python
#importing the dataset
import pandas as pd
import math
%matplotlib inline

data = pd.read_csv('data.csv')

#showing first five rows of data (before transformation)
data.head(5)
```

```python
#data transformations

def day_column():
    day = []
    for index, row in data.iterrows():
        if(row['day_Monday'] == 1):
            day.append("Monday")
        elif (row['day_Tuesday'] == 1):
            day.append("Tuesday")
        elif (row['day_Wednesday'] == 1):
            day.append("Wednesday")
        elif (row['day_Thursday'] == 1):
            day.append("Thursday")
        elif (row['day_Saturday'] == 1):
            day.append("Saturday")
        elif (row['day_Sunday'] == 1):
            day.append("Sunday")
    remove_cols = ['day_Monday', 'day_Tuesday', 'day_Wednesday', 'day_Thursday', 'day_Saturday', 'day_Sunday']
    data.drop(columns=remove_cols, inplace=True)
    data["day"] = day
data.head(5)
```

```python
def quarter_column():
    quarter = []
    for index, row in data.iterrows():
        if(row['quarter_Quarter1'] == 1):
            quarter.append(1)
        elif(row['quarter_Quarter2'] == 1):
            quarter.append(2)
        elif(row['quarter_Quarter3'] == 1):
            quarter.append(3)
        elif(row['quarter_Quarter4'] == 1):
            quarter.append(4)
        elif(row['quarter_Quarter5'] == 1):
            quarter.append(5)
    
    col_remove = ['quarter_Quarter1', 'quarter_Quarter2', 'quarter_Quarter3', 'quarter_Quarter4', 'quarter_Quarter5']
    data.drop(columns=col_remove, inplace=True)
    data['quarter_num'] = quarter
data.head(5)
```

```python
def department_column():
    department = []
    for index,row in data.iterrows():
        if (row['department_sweing'] == 1):
            department.append('sweing')
        else:
            department.append('finishing')
    
    col_remove = ['department_sweing', 'department_finishing', 'department_finishing']
    data.drop(columns=col_remove, inplace=True)
    data['department'] = department
data.head(5)
```

```python
def add_idle_column():
    idle_time = []
    idle_men = []
    for item,row in data.iterrows():
        if(row['idle_men'] != 0):
            idle_men.append(True)
        else:
            idle_men.append(False)
        if(row['idle_time'] != 0.0):
            idle_time.append(True)
        else:
            idle_time.append(False)
    data['idle_time?'] = idle_time
    data['idle_men?'] = idle_men
data.head(5)
```

```python
def num_workers_column():
    data['no_of_workers'] = data['no_of_workers'].apply(lambda x: math.ceil(x))

def wip_column():
    data['wip'] = data['wip'].fillna(0)

def remove_month_column():
    data.drop(columns='month', inplace=True)
data.head(5)
```

```python
#make the transformations and read into new csv

day_column()   
quarter_column()
department_column()  
add_idle_column()  
num_workers_column()
wip_column()
remove_month_column()
data.head(5)

new_csv_file = 'transformed_data.csv'

# Save the transformed data to a new CSV file
data.to_csv(new_csv_file, index=False)
```
